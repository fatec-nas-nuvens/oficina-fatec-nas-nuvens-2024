```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
from pyspark.sql.functions import row_number, monotonically_increasing_id 
from pyspark.sql.window import Window
```

```python
df = spark.read.format("delta").load("Tables/Vendas")

display(df)
```

##### **Criando a dimensão de "Categoria"**

```python
categoria = df.select("categoria").distinct()

dim_categoria = (categoria.withColumn("sk_categoria", monotonically_increasing_id() + 1))

dim_categoria.show()

dim_categoria.write.format("delta").mode("overwrite").save("abfss://85718698-872f-4ed2-b22e-50273ba4f3aa@onelake.dfs.fabric.microsoft.com/ffd6f5f5-79e2-4aba-bb00-cfaae220e4d4/Tables/dim_categoria")
```

##### **Criando a dimensão de "Subcategoria"**

```python
subcategorias = df.select("subcategoria").distinct()

dim_subcategoria = (subcategorias.withColumn("sk_subcategoria", monotonically_increasing_id() + 1))  # Adiciona + 1 para iniciar em 1

dim_subcategoria.show()

dim_subcategoria.write.format("delta").mode("overwrite").save("abfss://85718698-872f-4ed2-b22e-50273ba4f3aa@onelake.dfs.fabric.microsoft.com/ffd6f5f5-79e2-4aba-bb00-cfaae220e4d4/Tables/dim_subcategoria")
```

```python
prefato = df.join(dim_categoria, df.categoria == dim_categoria.categoria, 'left')

prefato = prefato.join(dim_subcategoria, prefato.subcategoria == dim_subcategoria.subcategoria, 'left')

prefato = prefato.join(dim_modelo, prefato.modelo == dim_modelo.modelo, 'left')

prefato = prefato.join(dim_cor, prefato.cor == dim_cor.cor, 'left')

prefato = prefato.join(dim_tamanho, prefato.tamanho == dim_tamanho.tamanho, 'left')

prefato = prefato.join(dim_uf, prefato.uf == dim_uf.uf, 'left')

prefato = prefato.join(dim_estado, prefato.estado == dim_estado.estado, 'left')

prefato = prefato.join(dim_cidade, prefato.cidade == dim_cidade.cidade, 'left')

prefato = prefato.join(dim_regiao, prefato.regiao == dim_regiao.regiao, 'left')

fato_vendas = prefato.select("id_venda", "sk_categoria", "sk_subcategoria","sk_modelo", "sk_cor", "sk_tamanho", "sk_uf", "sk_estado", "sk_cidade", "sk_regiao","data_pedido", "preco_custo_unit","pedidos_qtd", "total_venda")

fato_vendas.write.format("delta").mode("overwrite").save("abfss://85718698-872f-4ed2-b22e-50273ba4f3aa@onelake.dfs.fabric.microsoft.com/ffd6f5f5-79e2-4aba-bb00-cfaae220e4d4/Tables/fato_vendas")

display(fato_vendas)
```
