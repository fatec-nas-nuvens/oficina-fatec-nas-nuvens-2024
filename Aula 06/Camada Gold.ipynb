```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
from pyspark.sql.functions import row_number, monotonically_increasing_id 
from pyspark.sql.window import Window
```

```python
df = spark.read.format("delta").load("Tables/Vendas")

display(df)
```

##### **Criando a dimensão de "Categoria"**

```python
categoria = df.select("categoria").distinct()

dim_categoria = (categoria.withColumn("sk_categoria", monotonically_increasing_id() + 1))

dim_categoria.show()

dim_categoria.write.format("delta").mode("overwrite").save("abfss://85718698-872f-4ed2-b22e-50273ba4f3aa@onelake.dfs.fabric.microsoft.com/ffd6f5f5-79e2-4aba-bb00-cfaae220e4d4/Tables/dim_categoria")
```

##### **Criando a dimensão de "Subcategoria"**

```python
subcategorias = df.select("subcategoria").distinct()

dim_subcategoria = (subcategorias.withColumn("sk_subcategoria", monotonically_increasing_id() + 1))  # Adiciona + 1 para iniciar em 1

dim_subcategoria.show()

dim_subcategoria.write.format("delta").mode("overwrite").save("abfss://85718698-872f-4ed2-b22e-50273ba4f3aa@onelake.dfs.fabric.microsoft.com/ffd6f5f5-79e2-4aba-bb00-cfaae220e4d4/Tables/dim_subcategoria")
```

##### **Criando a dimensão de "Modelo"**

```python
modelo = df.select("modelo").distinct()

dim_modelo = (modelo.withColumn("sk_modelo", monotonically_increasing_id() + 1))  # Adiciona + 1 para iniciar em 1

dim_modelo.show()

dim_modelo.write.format("delta").mode("overwrite").save("abfss://85718698-872f-4ed2-b22e-50273ba4f3aa@onelake.dfs.fabric.microsoft.com/ffd6f5f5-79e2-4aba-bb00-cfaae220e4d4/Tables/dim_modelo")
```

##### **Criando a dimensão de "Cor"**

```python
cor = df.select("cor").distinct()

dim_cor = (cor.withColumn("sk_cor", monotonically_increasing_id() + 1))

dim_cor.show()

dim_cor.write.format("delta").mode("overwrite").save("abfss://85718698-872f-4ed2-b22e-50273ba4f3aa@onelake.dfs.fabric.microsoft.com/ffd6f5f5-79e2-4aba-bb00-cfaae220e4d4/Tables/dim_cor")
```
